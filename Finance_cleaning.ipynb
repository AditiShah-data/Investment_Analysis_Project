```python

# 1. Remove leading/trailing spaces from column names
df.columns = df.columns.str.strip()

# 2. Convert column names to lowercase
df.columns = df.columns.str.lower()

# 3. Remove extra spaces from string values in all object columns
df = df.apply(lambda col: col.str.strip() if col.dtype == 'object' else col)

# Optional: replace multiple internal spaces with single space
df = df.apply(
    lambda col: col.str.replace(r'\s+', ' ', regex=True).str.strip()
    if col.dtype == 'object' else col
)

```
